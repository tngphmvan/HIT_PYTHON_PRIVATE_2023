```python
# @title Default title text
#Bai1:

class Pycon:
  id = 0
  nextID = 1
  objects = []

  def __init__(self,ten = None, tuoi = None):
    self.ten = ten
    self.tuoi = tuoi
    self.id = Pycon.nextID
    Pycon.nextID += 1


  def nhap(self):
    self.ten = input()
    self.tuoi = int(input())

  @classmethod
  def avg(cls):
    tong_tuoi = 0
    for py in cls.objects:
      tong_tuoi += py.tuoi
    return tong_tuoi/len(cls.objects)

  @staticmethod
  def main():
    n = int(input())
    for i in range(n):
      py = Pycon()
      py.nhap()
      py.objects.append(py)
    for pycon in Pycon.objects:
      print(f"ID: {pycon.id} || Tên: {pycon.ten} || Tuổi: {pycon.tuoi}")
    print(f"Trung bình tuổi: {py.avg()}")

if __name__ == "__main__":
  Pycon.main()
```

```python
class Matrix:

    def __init__(self, n, m, value):
        self.n = n
        self.m = m
        self.value = value

    def __str__(self):
        return f"Matrix({self.n}, {self.m}): \n{self.value}"

    def __add__(self, other):
        if self.n != other.n or self.m != other.m:
            raise ValueError("Ma trận không cùng kích thước")
        return Matrix(self.n, self.m, [[self.value[i][j] + other.value[i][j] for j in range(self.m)] for i in range(self.n)])

    def __sub__(self, other):
        if self.n != other.n or self.m != other.m:
            raise ValueError("Ma trận không cùng kích thước")
        return Matrix(self.n, self.m, [[self.value[i][j] - other.value[i][j] for j in range(self.m)] for i in range(self.n)])

    def __mul__(self, other):
        if self.m != other.n:
            raise ValueError("Số cột của ma trận thứ nhất không bằng số hàng của ma trận thứ hai")
        return Matrix(self.n, other.m, [[sum([self.value[i][k] * other.value[k][j] for k in range(self.m)]) for j in range(other.m)] for i in range(self.n)])

    def __truediv__(self, other):
        if other.n != self.n or other.m != self.m:
            raise ValueError("Ma trận không cùng kích thước")
        return Matrix(self.n, self.m, [[self.value[i][j] / other.value[i][j] for j in range(self.m)] for i in range(self.n)])

    def __eq__(self, other):
        if self.n != other.n or self.m != other.m:
            return False
        for i in range(self.n):
            for j in range(self.m):
                if self.value[i][j] != other.value[i][j]:
                    return False
        return True

def main():
  A = Matrix(2, 2, [[1, 2], [3, 4]])
  B = Matrix(2, 2, [[5, 6], [7, 8]])
  print(A)
  print(B)

if __name__ == "__main__":
  main()
```
