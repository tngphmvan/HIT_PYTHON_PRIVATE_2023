```python
#Bai 6:
def tinh_loss(x: list[float], y: list[float], w: float, b: float) -> float:
    """
    Hàm tính chi phí Loss.

    Args:
        x: Danh sách diện tích căn hộ.
        y: Danh sách giá thực tế của căn hộ.
        w: Hệ số w.
        b: Hệ số b.

    Returns:
        Chi phí Loss.
    """

    c = []
    for i in range(len(x)):
        c.append(w * x[i] + b)
    sum = 0
    for i in range(len(x)):
        sum += (c[i] - y[i])**2
    return 1 / (2 * len(x)) * sum

def tim_w_va_b_tot_min(x: list[float], y: list[float]):
    """
    Hàm tìm w và b sao cho Loss nhỏ nhất.

    Args:
        x: Danh sách diện tích căn hộ.
        y: Danh sách giá thực tế của căn hộ.

    Returns:
        Hệ số w và b.
    """

    w_min = 0
    b_min = 0
    loss_min = float("inf")
    for w in range(-10, 10):
        for b in range(-100, 100):
            loss = tinh_loss(x, y, w / 10, b)
            if loss < loss_min:
                loss_min = loss
                w_min = w
                b_min = b
    return w_min, b_min

def du_doan_gia_can_ho(x: float, w: float, b: float) -> float:
    """
    Hàm dự đoán giá căn hộ.

    Args:
        x: Diện tích căn hộ.
        w: Hệ số w.
        b: Hệ số b.

    Returns:
        Giá căn hộ dự đoán.
    """

    return w * x + b

if __name__ == "__main__":
    x = [50, 75, 80, 60, 70,90,100,65,85,105]
    y = [600,860,930,700,800,1040,1150,750,970,1200]

    w, b = tim_w_va_b_tot_min(x, y)
    print("Hệ số w và b tối ưu:", w, b)

    # Dự đoán giá căn hộ có diện tích 40 m2
    gia_du_doan = du_doan_gia_can_ho(40, w, b)
    print("Giá căn hộ có diện tích 40 m2:", gia_du_doan)
```
